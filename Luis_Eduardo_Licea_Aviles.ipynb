```python
#Gracias a la opcion info() podemos notar que hay algunos valores nulos en el dataframe de productos
#Reemplazamos los valores nulos por la palabra N/A
productos.fillna("N/A",inplace=True)
```

```python
# Unir los DataFrames
#unimos segmento con categoria
segcat = pd.merge(categoria, segmentos, how='inner', left_on='ID_CATEGORY', right_on='CATEGORY')
# Eliminar la columna redundante id_categoria
segcat = segcat.drop('CATEGORY_y', axis=1)
```

```python
#unimos los productos con las ventas
ventasyproductos=pd.merge(ventas,productos, how='inner', left_on='ITEM_CODE', right_on='ITEM')
#Eliminamos la columna redundate Item_Code
ventasyproductos = ventasyproductos.drop('ITEM', axis=1)
```

```python
#Añadimos una columna extra para calculos y análisis
#Creamos una columna para saber el precio promedio por unidad, dividimos total_values_sales/total_unit sales
#Nos aseguramos de no dividir entre 0
import numpy as np
ventasyproductos["Price_Per_unit"]=np.where(
    ventasyproductos["TOTAL_UNIT_SALES"]!=0,
    ventasyproductos["TOTAL_VALUE_SALES"]/ventasyproductos["TOTAL_UNIT_SALES"],
    0
)
```

```python
#Guardamos los dataframes para su analisis posterior en un archivo de excel
dataframes = {
    "Ventas": ventasyproductos,
    "Categorias": segcat,
    "Productos": productos,
    "Calendario": calendario
}
with pd.ExcelWriter("Concentrado_de_Ventas.xlsx") as writer:
    for sheet_name, df in dataframes.items():
        df.to_excel(writer, sheet_name=sheet_name, index=False)
```

```python
#CREACION DE GRAFICOS
#GRAFICO PARA MOSTRAR LA DISTRIBUCION DE VENTAS POR FORMATO (Ya que todas las ventas son de la categoria 1, se optó por analizarlas de acuerdo al formato)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(data=ventasyproductos, x='FORMAT', y='TOTAL_UNIT_SALES')

# Personalizar el gráfico
plt.title('Distribución de Ventas por Formato (Boxplot)', fontsize=14)
plt.xlabel('Formato', fontsize=12)
plt.ylabel('Ventas Totales (Unidades)', fontsize=12)

# Rotar etiquetas si hay muchos formatos
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.show()
```

```python
#Ventas por Region
# Agrupar los datos por WEEK y REGION, sumando TOTAL_UNIT_SALES
ventas_por_region = ventasyproductos.groupby(['WEEK', 'REGION'])['TOTAL_UNIT_SALES'].sum().reset_index()

# Crear el gráfico de líneas
plt.figure(figsize=(12, 6))
sns.lineplot(data=ventas_por_region, x='WEEK', y='TOTAL_UNIT_SALES', hue='REGION', marker='o')

# Personalizar el gráfico
plt.title('Ventas Totales por Región a lo Largo del Tiempo', fontsize=14)
plt.xlabel('Semana', fontsize=12)
plt.ylabel('Ventas Totales (Unidades)', fontsize=12)
plt.legend(title='Región')

# Rotar etiquetas si hay muchas semanas
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.show()
```

```python
#Grafico de los productos más vendidos
# Identificar los 6 productos más vendidos (por suma total de TOTAL_UNIT_SALES)
top_6_productos = ventasyproductos.groupby('ITEM_CODE')['TOTAL_UNIT_SALES'].sum().nlargest(6).index

# Filtrar ventasyproductos para incluir solo los 6 productos más vendidos
ventas_top_6 = ventasyproductos[ventasyproductos['ITEM_CODE'].isin(top_6_productos)]

# Crear un diccionario para mapear ITEM_CODE a ITEM_DESCRIPTION
# Buscar las descripciones en el DataFrame productos
descripcion_map = productos[productos['ITEM'].isin(top_6_productos)][['ITEM', 'ITEM_DESCRIPTION']].set_index('ITEM')['ITEM_DESCRIPTION'].to_dict()

# Agregar una columna ITEM_DESCRIPTION a ventas_top_6 usando el mapeo
ventas_top_6 = ventas_top_6.copy()  # Evitar advertencias de modificación
ventas_top_6['ITEM_DESCRIPTION'] = ventas_top_6['ITEM_CODE'].map(descripcion_map)

# Agrupar por WEEK y ITEM_DESCRIPTION, sumando TOTAL_UNIT_SALES
ventas_por_semana = ventas_top_6.groupby(['WEEK', 'ITEM_DESCRIPTION'])['TOTAL_UNIT_SALES'].sum().reset_index()

# Crear el gráfico de líneas
plt.figure(figsize=(12, 6))
sns.lineplot(data=ventas_por_semana, x='WEEK', y='TOTAL_UNIT_SALES', hue='ITEM_DESCRIPTION', marker='o')

# Personalizar el gráfico
plt.title('Ventas de los 6 Productos Más Vendidos por Semana', fontsize=14)
plt.xlabel('Semana', fontsize=12)
plt.ylabel('Ventas Totales (Unidades)', fontsize=12)
plt.legend(title='Descripción del Producto', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

# Rotar etiquetas si hay muchas semanas
plt.xticks(rotation=45)

# Ajustar el diseño para evitar que la leyenda se corte
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
#Ventas por region
# Agrupar los datos por WEEK y REGION, sumando TOTAL_UNIT_SALES
ventas_por_region = ventasyproductos.groupby(['WEEK', 'REGION'])['TOTAL_UNIT_SALES'].sum().reset_index()

# Pivotar los datos para tener REGION como columnas y WEEK como índice
ventas_pivot = ventas_por_region.pivot(index='WEEK', columns='REGION', values='TOTAL_UNIT_SALES').fillna(0)

# Crear el gráfico de barras apiladas
plt.figure(figsize=(12, 6))
ventas_pivot.plot(kind='bar', stacked=True, ax=plt.gca())

# Personalizar el gráfico
plt.title('Ventas Totales por Región (Barras Apiladas)', fontsize=14)
plt.xlabel('Semana', fontsize=12)
plt.ylabel('Ventas Totales (Unidades)', fontsize=12)
plt.legend(title='Región', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)

# Ajustar el diseño para evitar que la leyenda se corte
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
#Productos más vendidos
# Filtrar  para reducir el número de productos (los 20 más vendidos)
top_20_productos = ventasyproductos.groupby('ITEM_CODE')['TOTAL_UNIT_SALES'].sum().nlargest(20).index
ventas_filtrado = ventasyproductos[ventasyproductos['ITEM_CODE'].isin(top_20_productos)]

# Crear boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=ventas_filtrado, x='ITEM_CODE', y='TOTAL_UNIT_SALES')
plt.title('Distribución de Ventas por Producto (Top 20)', fontsize=14)
plt.xlabel('Código del Producto', fontsize=12)
plt.ylabel('Ventas Totales (Unidades)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#por categoria
plt.figure(figsize=(10, 6))
sns.boxplot(data=ventasyproductos, x='CATEGORY', y='TOTAL_UNIT_SALES')
plt.title('Distribución de Ventas por Categoría', fontsize=14)
plt.xlabel('Categoría', fontsize=12)
plt.ylabel('Ventas Totales (Unidades)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Ventas por region
plt.figure(figsize=(10, 6))
sns.boxplot(data=ventasyproductos, x='REGION', y='TOTAL_UNIT_SALES')
plt.title('Distribución de Ventas por Región', fontsize=14)
plt.xlabel('Región', fontsize=12)
plt.ylabel('Ventas Totales (Unidades)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(data=ventasyproductos, x='TOTAL_UNIT_SALES', y='Price_Per_unit', alpha=0.6)

# Personalizar el gráfico
plt.title('Ventas Totales (Unidades) vs Precio por Unidad', fontsize=14)
plt.xlabel('Ventas Totales (Unidades)', fontsize=12)
plt.ylabel('Precio por Unidad', fontsize=12)

# Ajustar el diseño
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
